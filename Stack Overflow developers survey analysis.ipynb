```python
""" Creating a function which implements one hot encoding technique to convert a categorical column to a dataframe 
with multiple columns each representing a category with values True and False. """
```

```python
def split_multicolumn(col_series):
    # Convert the Series to a DataFrame
    result_df = col_series.to_frame()
    options = []
    
    # Iterate over the non-null elements of the Series
    for index, value in col_series[col_series.notnull()].items():
        # Split each value into a list of options
        for option in value.split(';'):
            # If the option is not already a column, add it
            if option not in result_df.columns:
                options.append(option)
                result_df[option] = False
            # Mark the value in the option column as True
            result_df.at[index, option] = True
    
    # Return the DataFrame with only the one-hot encoded columns
    return result_df[options]
```

## Most popular languages developers worked with

```python
languages_worked_df = split_multicolumn(df.LanguageHaveWorkedWith)
```

```python
languages_worked_df_percentage = languages_worked_df.mean().sort_values(ascending=False) * 100
languages_worked_df_percentage.head(10)
```

```python
plt.figure(figsize=(12, 16))
plt.xlabel("Percent %")
sns.barplot(x=languages_worked_df_percentage, y=languages_worked_df_percentage.index)
plt.title("Most used languages by developers");
```

## Languages people want to learn and work with in the future
